```python
import numpy as np
from numpy import float32
from numpy import float64
import matplotlib.pyplot as plot
from math import sin
from math import log

TERMS_NUM = 1000001
TERMS_NUM_MODEL = 709
```

```python
def calculate_nth_term(n, z):
    return 1.0 / (n ** 2 - n - z)

def calculate_nth_term_model(n, z):
    return (z + n) / ((n ** 2 - n - z) * (n ** 2))

def calculate_sum(z):
    part_sum = 0
    for i in range(0, TERMS_NUM):
        part_sum += calculate_nth_term(i + 1, z)
    return part_sum

def calculate_model_sum(z):
    part_sum = 0
    for i in range(0, TERMS_NUM_MODEL):
        part_sum += calculate_nth_term_model(i + 1, z)
    return 1.644934 + part_sum
```

```python
print(calculate_sum(1.5))
print(calculate_model_sum(1.5))
```

```python
plot.figure(figsize=(15, 10))
interval = np.arange(0.1, 2, 0.1)
plot.plot(interval, [calculate_sum(i) for i in interval])
plot.show()
plot.figure(figsize=(15, 10))
plot.plot(interval, [calculate_model_sum(i) for i in interval])
plot.show()
plot.figure(figsize=(15, 10))
plot.plot(interval, [abs(calculate_sum(i) - calculate_model_sum(i)) for i in interval])
plot.show()
```
