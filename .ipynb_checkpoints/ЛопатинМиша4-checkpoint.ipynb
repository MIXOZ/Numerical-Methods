## Task C
Chebish nodes

````python
def get_chebish_nodes(n):
    return [cos((pi * (2 * k - 1)) / (2 * n)) for k in range(1, n + 1)]
````

````python
def get_local_chebish(start, stop, n):
    chebish_nodes = np.array(get_chebish_nodes(n))
    chebish_nodes = chebish_nodes - np.array([-1] * n)
    chebish_nodes = chebish_nodes * np.array([(stop - start) / 2] * n)
    return chebish_nodes + np.array([start] * n)
````

````python
def generate_chebish_nodes(func, x_0, eps, num, rnd=True):
    x_arr = get_local_chebish(x_0 - eps, x_0 + eps, num)
    y_arr = [func(x) for x in x_arr]
    return x_arr, y_arr
````

````python
def print_max_error_chebish_nodes(func=FUNC, eps=EPS, x_0=X_0):
    plt.figure(figsize=(20, 5))
    max_error_arr = []
    for n in range(5, 51):
        x_arr, y_arr = generate_chebish_nodes(func, x_0, eps, n, False)
        points_x_arr = np.arange(x_0 - eps, x_0 + eps, 2 * eps / 1000)
        points_y_arr_lagrange = np.array(calculate_interpolate_lagrange_multiple_points(points_x_arr, x_arr, y_arr))
        points_y_arr_actual = np.array([func(x) for x in points_x_arr])
        error_arr = np.abs(points_y_arr_lagrange - points_y_arr_actual)
        max_error_arr.append(np.max(error_arr))
    plt.title("Max error Chebish nodes")
    plt.plot(range(5, 51), max_error_arr)
    plt.show()        
````
